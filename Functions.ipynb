```python
#factorial function
# Adarsh Gurjar
def fact(n):
    a=1
    for i in range(1, n + 1):
        a= a* i
    return a
print(fact(5))
```

```python
# checks no. is Prime or not
# Adarsh Gurjar
def prime(n):
    count = 0

    for i in range(1, n + 1):
        if n % i == 0:
            count += 1

    if count == 2:
        print("Prime")
    else:
        print("Not Prime")

prime(11)
```

```python
# prints star pattern
# Adarsh Gurjar
def pattern(rows):
    for i in range(1, rows + 1):
        print("*" * i)

pattern(7)
```

```python
# prints diamond star pattern
# Adarsh Gurjar
def diamond(rows):
    for i in range(1, rows + 1):
        print(" " * (rows - i) + "*" * (2 * i - 1))
    for i in range(rows - 1, 0, -1):
        print(" " * (rows - i) + "*" * (2 * i - 1))

diamond(5)
```

```python
# checks palindrome string
# adarsh gurjar
def palindrome(text):
    if text == text[::-1]:
        print("Palindrome")
    else:
        print("Not Palindrome")

palindrome("asdsa")
```

```python
# prints multiplication table
def table(n):
    for i in range(1, 11):
        print(n, "x", i, "=", n*i)

table(5)
```

```python
# fibonacci series
def fibonacci(n):
    a = 0
    b = 1
    for i in range(n):
        print(a, end=" ")
        c = a + b
        a = b
        b = c

fibonacci(10)
```
